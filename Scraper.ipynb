```python
import time
import random
import pandas as pd
import sqlite3
import requests
```

```python
pornlinks_filename  = './links/porn_links/porn_links.txt'
url_25k_filename = './links/url_25k.tsv'
dmoz_url_filename = './links/dmoz/urls_dmoz.csv'
```

```python
def save_dataset(df,name):
    df.to_csv(name,index=False)
    print(f"successfully saved {name}")
```

# Porn links dataset

```python
# Using readlines()
file1 = open(pornlinks_filename, 'r')
lines = file1.readlines()

lines = [line.replace('\n','') for line in lines]

df = pd.DataFrame(columns=['url','category'])

df['url'] = lines
df['category'] = 'Porn'

save_dataset('porn_urls.csv')
```

# url_25k dataset

```python
df = pd.read_csv(url_25k_filename, sep='\t', on_bad_lines='skip')
df = df[["URL","Primary Category"]]
df.rename(columns={'URL':"url","Primary Category":"category"},inplace=True)
save_dataset(df,"url_25k.csv")
```
