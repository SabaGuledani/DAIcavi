```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt 
import pandas as pd
from random import randrange
import time
```

```python
df = pd.read_csv('../datasets/images/memegenerator.csv')
meme_ids = df["Meme ID"].values
meme_ids = list(meme_ids)
```

```python
img_idx = 2000
for id in meme_ids[2000:]:
    res = requests.get(f"https://img.memegenerator.net/instances/600x600/{id}.jpg")
    if res.status_code == 200:
        try:
            image = Image.open(BytesIO(res.content))
            image.save(f"../datasets/images/memegenerator_images/{id}.jpg")
            interval = randrange(100,250)/100
            time.sleep(interval)
            print(img_idx)
            
        except:
            print("Failed to open")
    else:
        print(f"Failed to retrieve image. Status code: {res.status_code}")
    img_idx+=1
    
```
